```python
import pandas as pd
df=pd.read_csv("six_sector.unknown")
# df["Entry"]=df["Entry"].astype(str)
df
# df=pd.to_datetime(df["Entry"],format="%m/%y",dayfirst=True)
# df
```

```python

fintech=df["Sector"].str.contains("Fintech")
df_fintech=df[fintech]
df_fintech["Fintech"]="Fintech"

ecom=df["Sector"].str.contains("E-Commerce")
df_ecom=df[ecom]
df_ecom["E-Commerce"]="E-Commerce"

health=df["Sector"].str.contains("Healthtech")
df_health=df[health]
df_health["Healthtech"]="Healthtech"

saas=df["Sector"].str.contains("SaaS")
df_saas=df[saas]
df_saas["Sector"]="SaaS"

edtech=df["Sector"].str.contains("Edtech")
df_edtech=df[edtech]
df_edtech["Sector"]="Edtech"

market=df["Sector"].str.contains("Marketplace")
df_market=df[market]
df_market["Sector"]="Marketplace"
df_fintech




```

```python
six_sector=pd.concat(objs=[df_fintech,df_ecom,df_health,df_saas,df_market,df_edtech])
six_sector.to_csv("six_sector")
```
